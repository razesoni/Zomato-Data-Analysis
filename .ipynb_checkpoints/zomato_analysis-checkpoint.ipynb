### Zomato Dataset Consist the following columns :

+ <small><b>Restaurant Id:</b> Unique id of every restaurant across various cities of the world
+ <small><b>Restaurant Name:</b> Name of the restaurant
+ <small><b>Country Code:</b> Country in which restaurant is located
+ <small><b>City:</b> City in which restaurant is located
+ <small><b>Address:</b> Address of the restaurant
+ <small><b>Locality:</b> Location in the city
+ <small><b>Locality Verbose:</b> Detailed description of the locality
+ <small><b>Longitude:</b> Longitude coordinate of the restaurant's location
+ <small><b>Latitude:</b> Latitude coordinate of the restaurant's location
+ <small><b>Cuisines:</b> Cuisines offered by the restaurant
+ <small><b>Average Cost for two:</b> Cost for two people in different currencies
+ <small><b>Currency:</b> Currency of the country
+ <small><b>Has Table booking:</b> yes/no
+ <small><b>Has Online delivery:</b> yes/ no
+ <small><b>Is delivering:</b> yes/ no
+ <small><b>Switch to order menu:</b> yes/no
+ <small><b>Price range:</b> range of price of food
+ <small><b>Aggregate Rating:</b> Average rating out of 5
+ <small><b>Rating color:</b> depending upon the average rating color
+ <small><b>Rating text:</b> text on the basis of rating of rating
+ <small><b>Votes:</b> Number of ratings casted by people

### Loading the Country-Code Dataset:

```python
 df_country = pd.read_csv("Country-Code.csv")
```

```python
df_country
```

### Merging both the Datasets :

```python
final_df = pd.merge(df, df_country, on="Country Code", how="left"  )
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
country_names = list(set(final_df['Country']))
print("Here is the list of all the countries that uses zomato: ")
for i in country_names:
    print(i)
```

```python
country_list = final_df['Country'].value_counts()
```

```python
print("Here are the list of transaction happened in each countries:")
country_list
```

<b>Observation: </b>Maximum Transactions happened in India 8652

```python
plt.pie(country_list.values[:3], labels=country_list.index[:3], autopct='%.2f%%')
```

The above pie chart shows the top 3 countries with maximum transactions with percentage

```python
# Analyzing average cost for two people per country
final_df.groupby('Country')['Average Cost for two'].mean().reset_index()
```

### Let's see how Has Online delivery and Has Table booking affects Aggregate rating:

```python
final_df.groupby(['Country', 'Has Online delivery']).size().reset_index()
```

<b>Observation: </b>Only India and UAE accepts online delivery

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Has Online delivery', y='Aggregate rating', data=final_df, palette=['skyblue', 'pink'])
plt.title('Has Online delivery vs Aggregate rating')
```

```python
final_df.groupby(['Country', 'Has Table booking']).size().reset_index()
```

<b>Observation :</b>India, Phillipines, Qatar, South Africa, UAE and United Kingdom have option for table booking

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Has Table booking', y='Aggregate rating', data=final_df, palette=['skyblue', 'pink'])
plt.title('Has Table booking vs Aggregate rating')
```

```python
final_df.columns
```

```python
rating = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0 :'Rating count'})
```

```python
rating
```

<b>Observation:</b><br>
When Rating is between 4.5 to 4.9 it means Excellent<br>
When Rating is between 4.0 to 4.4 it means Very Good<br>
When Rating is between 3.5 to 3.9 it means Good<br>
When Rating is between 2.5 to 3.4 it means Average<br>
When Rating is between 2.0 to 2.4 it means Poor

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating', y ='Rating count', data=rating, hue='Rating color', palette='magma')
```

<b>Observation: </b><br>1. Not Rated count is very high<br>
                    2. The above graph shows that the majority ratings are done between 2.8 to 3.9

```python
total_votes_by_country = final_df.groupby('Country')['Votes'].sum().reset_index()
total_votes_by_country
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x ='Aggregate rating', y ='Votes', data=final_df)
plt.title('Relationship between Aggregeate rating and Votes')
```

<b>Observation: </b><br>
1. The Aggregate rating increases with the no. of votes
2. Many people have not voted

```python
votes_by_restaurant = final_df.groupby('Restaurant Name')['Votes'].sum().reset_index()
```

```python
most_voted_restaurant = votes_by_restaurant.sort_values(by='Votes', ascending=False)[:10].reset_index()
```

```python
print('Here are the top 10 most voted restaurant:')
most_voted_restaurant
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Restaurant Name', y='Votes', data=most_voted_restaurant)
plt.xticks(rotation = 90)
plt.title('Top 10 most voted restaurant')
```

```python
rate_restaurant = final_df.groupby('Restaurant Name')['Aggregate rating'].agg(['mean', 'count']).reset_index()
```

```python
restaurant_mean = rate_restaurant.merge(votes_by_restaurant, on='Restaurant Name')
```

```python
restaurant_mean
```

```python
restaurant_mean = restaurant_mean[restaurant_mean['Votes'] > 1000]
```

```python
highest_rated_restaurant = restaurant_mean.sort_values(by='mean', ascending=False)[:10]
```

```python
print('Here are the top 10 highest rated restaurant with having more than 1000 votes:')
highest_rated_restaurant
```

```python
cuisines_df = final_df['Cuisines'].str.split(',').explode()
```

```python
cuisines_df
```

```python
top_cuisines = cuisines_df.value_counts()[:15].reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='Cuisines', data=top_cuisines, palette='inferno')
plt.title('top 15 cuisines')
```

<b>Observation: </b>Most people like to order North Indian cuisines in the restaurant followed by Chinese and Fast Food

```python
top_cities = final_df['City'].value_counts().head(10).reset_index()
```

```python
print("Here are the list of top 10 cities in the Dataset:")
top_cities
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='City', data=top_cities)
plt.title("Top 10 Cities by no. of Restaurants")
plt.xlabel('No. of Restaurants')
plt.ylabel('Cities')
```
